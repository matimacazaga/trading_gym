```python
import pickle
import numpy as np
import pandas as pd
import talib 
```

```python
random_universes = pickle.load(open("random_universes.pickle", "rb"))
```

```python
close = random_universes[0]["close"]
open = random_universes[0]["open"]
low = random_universes[0]["low"]
high = random_universes[0]["high"]
volume = random_universes[0]["volume"]
```

```python
momentum = close.apply(lambda x: talib.MOM(x, timeperiod=14))
momentum.tail()
```

```python
def get_macd(close:np.ndarray)->np.ndarray:

    macd, _, _ = talib.MACD(close, fastperiod=12, slowperiod=26)

    return macd

    
```

```python
macd = close.apply(get_macd)
macd.tail()
```

```python
mfi = close.apply(lambda x: talib.MFI(high.loc[x.index, x.name], low.loc[x.index, x.name], x, volume.loc[x.index, x.name]))
mfi.tail()
```

```python
natr = close.apply(lambda x: talib.NATR(high.loc[x.index, x.name], low.loc[x.index, x.name], x))
natr.tail()
```

```python
htdcp = close.apply(talib.HT_DCPHASE)
htdcp.tail()
```

```python
def get_ht_sine(close:np.ndarray):

    sine, _ = talib.HT_SINE(close)

    return sine
```

```python
hts = close.apply(get_ht_sine)
hts.tail()
```

```python
httmm = close.apply(talib.HT_TRENDMODE)
httmm.tail()
```

```python
obv = close.apply(lambda x: talib.OBV(x, volume.loc[x.index, x.name]))
obv.tail()
```

```python
co = close.apply(lambda x: talib.ADOSC(high.loc[x.index, x.name], low.loc[x.index, x.name], x, volume.loc[x.index, x.name]))
co.tail()
```

```python
features_df = [("momentum", momentum), ("macd", macd), ("mfi", mfi), ("natr", natr), ("htdcp", htdcp), ("hts", hts), ("httmm", httmm), ("on_balance_vol", obv)]
df = co.stack().reset_index().rename({"level_1": "symbol", 0: "chaikin_oscillator"}, axis=1)
for feature_name, feature_df in features_df:
    tmp = feature_df.stack().reset_index().rename({"level_1": "symbol", 0: feature_name}, axis=1)
    df = df.merge(tmp, how="inner", on=["date", "symbol"])
df.head()
```

```python
df.shape
```

```python
from typing import Optional
def create_target(close:pd.DataFrame, n:int=1, target_type:str="binary", q:Optional[int]=None):
    target = (close.shift(-n)/close) - 1.
    target = target.stack().reset_index().rename({"level_1": "symbol", 0: "target"}, axis=1)
    if target_type == "binary":
        if q is None:
            target.loc[:, "target"] = target.loc[:, "target"].apply(lambda x: 1 if x>0 else 0)
        else:
            target.loc[:, "target"] = target.groupby("date").apply(lambda x: pd.qcut(x["target"], q=q, labels=False)).values
            target.loc[:, "target"] = target.loc[:, "target"].apply(lambda x: 1 if x == q-1 else 0)
    else:
        if q is None:
            raise ValueError("q must be provided if target_type='multi'")
        target.loc[:, "target"] = target.groupby("date").apply(lambda x: pd.qcut(x["target"], q=q, labels=False)).values
    
    return target
```

```python
target = create_target(close, target_type="multi", q=5)
```

```python
df.shape
```

```python
df = df.merge(target, on=["date", "symbol"], how="left")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.drop(["date", "symbol", "target"], axis=1))
pca = PCA(n_components=0.95)
X_scaled_pca = pca.fit_transform(X_scaled)
df_transformed = df.loc[:, ["date", "symbol", "target"]].copy()
df_transformed.loc[:, range(X_scaled_pca.shape[1])] = X_scaled_pca
df_transformed.head()

```

```python
import pywt
```

```python
haar = pywt.Wavelet("haar")
features = []

for name, group in df_transformed.groupby("symbol"):
    group_ = group.reset_index(drop=True)
    for c in df_transformed.drop(["date","symbol", "target"], axis=1).columns:
        ca, cd1, cd2 = pywt.wavedec(group_.loc[:, c].values, haar, mode="periodization", level=2)
        cd1 = np.where(cd1 >= 2.*cd1.std(ddof=1, axis=0), 0., cd1)
        cd2 = np.where(cd2 >= 2.*cd2.std(ddof=1, axis=0), 0., cd2)
        cd3 = np.where(cd3 >= 2.*cd3.std(ddof=1, axis=0), 0., cd3)
        denoised_signal = pywt.waverec([ca, cd1, cd2], haar)
        group_.loc[:, c] = denoised_signal[1:] if len(group_)%2 != 0 else denoised_signal
    features.append(group_)
```

```python
df_transformed_denoised = pd.concat(features, ignore_index=True).sort_values("date").dropna(axis=0, subset=["target"])
```

```python
df_transformed.dropna(axis=0, subset=["target"], inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

rf = RandomForestClassifier()
X = df_transformed.drop(["date", "symbol", "target"], axis=1).values
y = df_transformed.loc[:, "target"].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, rf.predict(X_test)))
```

```python
from trading_gym.envs.data_loader import CryptoData
from datetime import datetime 

start = datetime(2019, 5, 1)
end = datetime(2021, 9, 30)

cd = CryptoData()

supported_symbols = cd.supported_symbols

symbols_filtered = [
    symbol
    for symbol in supported_symbols
    if (
        symbol.endswith("USDT")
        or symbol.endswith("BUSD")
        or symbol.endswith("USDC")
    )
    and not symbol.startswith("USDT")
    and not symbol.startswith("USDS")
    and not symbol.startswith("BUSD")
    and not symbol.startswith("USDC")
    and not symbol.startswith("TUSD")
    and not symbol.startswith("DAI")
    and not symbol.startswith("PAX")
    and not symbol.startswith("EUR")
    and not symbol.startswith("MDX")
    and symbol not in ["UNIDOWNUSDT",      
"LTCDOWNUSDT",      
"SUSHIDOWNUSDT",    
"ETHDOWNUSDT",      
"BNBDOWNUSDT",      
"XLMDOWNUSDT",       
"SXPDOWNUSDT",       
"TRXDOWNUSDT",       
"YFIDOWNUSDT",       
"XTZDOWNUSDT",       
"AAVEDOWNUSDT",      
"DOTDOWNUSDT",       
"LINKDOWNUSDT",      
"ADADOWNUSDT",       
"XRPDOWNUSDT",        
"COCOSUSDT",]
]

symbols = {}

for symbol in symbols_filtered:

    if "USDT" in symbol:
        quote = "USDT"
    elif "BUSD" in symbol:
        quote = "BUSD"
    elif "USDC" in symbol:
        quote = "USDC"
    else:
        continue

    s = symbol.replace(quote, "")
    if s in symbols:
        continue
    else:
        symbols[s] = symbol

symbols = list(symbols.values())

print("Downloading data")

assets_data = cd.get_data(start=start, end=end, universe=symbols)

symbols = list(filter(lambda s: s in assets_data["close"].columns, symbols))
```

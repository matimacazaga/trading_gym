```python
from trading_gym.utils.summary import stats
```

```python
agent = "hrp"
start_eval_date = datetime(2019,1,1)

agents_statistics = []
for i in range(16):
    results = pickle.load(open(f"{path}/{agent}/results_{agent}_params_{i}_mix_weekly_reb_nocash_v2.pickle", "rb"))
    statistics = stats(results["agent"].rewards.loc[start_eval_date:].sum(axis=1))
    agents_statistics.append(
        {
            "window": results["params"]["window"],
            "n_obs": results["params"]["screener"].n_obs,
            "n_assets": results["params"]["screener"].n_assets,
            "Cumulative Return": statistics["Cumulative Returns"].iloc[-1],
            "Sharpe Ratio": statistics["Sharpe Ratio"]*np.sqrt(365.)
        }
    ) 
pd.DataFrame(agents_statistics).sort_values("Sharpe Ratio", ascending=False)
```
